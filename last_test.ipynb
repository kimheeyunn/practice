# DTM

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#max_features 최대 단어수 tokenizer -> 명사만 추출해서 dtm 작업을 하기 위해
vectorizer=CountVectorizer(max_features=1000,tokenizer=extract_noun)

DTM=vectorizer.fit_transform(data_list)
```

```python
dtm_df = pd.DataFrame(
    {
        'word':vectorizer.get_feature_names_out(),
        'dtm_sum':DTM.sum(axis=0).flat
    }
)
```

```python
def extract_noun(txt):
  result=kiwi.tokenize(txt)
  for token in result:
    if token.tag in ['NNG','NNP']:
      yield token.form
```

```python
from kiwipiepy import Kiwi
kiwi=Kiwi()
result=kiwi.tokenize("형태소분석결과입니다")
print(result)
```
